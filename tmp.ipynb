```python
import os
from pathlib import Path
import json
tot = 0
data_dir = Path("/mnt/ecf82360-d01d-4966-b234-c47ea01078db/datas1224")
failed = []

webs = os.listdir(data_dir)
total_html=0
total_pdf=0
for web in webs:
    try:
        print(f"{web}\nraw_html_count:{len(os.listdir(data_dir/web/'raw_htmls'))}")
        total_html+= len(os.listdir(data_dir/web/'raw_htmls'))
        # total_pdf+= len(os.listdir(data_dir/web/'raw_pdfs'))
        table = []
        with open(data_dir/web/'table.jsonl', 'r', encoding='utf-8') as f:
            for line in f.readlines():
                D = json.loads(line)
                table.append(D)
                if D['status'] == 'FAIL':
                    failed.append(D)
        print(f"{len(table)=}")
        urls = [i['url'] for i in table]
        print(f"{len(set(urls))=}")
        print(f"{len(failed)=}")
        print('-'*100)
    except:
        continue


print(f"{total_pdf=}\n{total_html=}")
print(f"{total_html+total_pdf=}")
tot += total_html+total_pdf
```
